```python
from snowflake.snowpark.session import Session
import os
```

```python
def create_sf_session():
    params = {
        "account": os.getenv("ID"),
        "user": os.getenv("NAME"),
        "password": os.getenv("PASS"),
        "role": os.getenv("ROLE"), 
    }
    session = Session.builder.configs(params).create()
    return session
```

```python
my_session = create_sf_session()
my_session.use_database('TEST')
# Your code goes here, inside the "main" handler.
tableName = 'TEST.PUBLIC.IRIS'
dataframe = my_session.table(tableName)

# Print a sample of the dataframe to standard output.
dataframe.show()
```
